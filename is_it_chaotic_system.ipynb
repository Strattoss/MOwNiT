```julia
using Plots

plot()
eps_t = [0.66666666666, 0.66666666667, 0.66666666668, 0.66666666669, 0.66666666670]

for eps in eps_t
    n = 500

    #x2_h = [1.0]
    #y2_h = [0.0]
    x1_h = [0.0]
    y1_h = [5.0]

    x2 = 1
    y2 = 0
    x1 = 0
    y1 = 5


    for iter in 1:n
        a = (y1 - y2) / (x1 - x2)
        b = y1 - a*x1
        
        x3 = x1 + sqrt(eps/(a*a + 1))
        y3 = a * x3 + b
        
        x1 = x3
        y1 = y3
        x2 =  x2 + eps
        
        push!(x1_h, x1)
        push!(y1_h, y1)
        #push!(y2_h, y2)
        #push!(x2_h, x2)
    
    end

    plot!(x1_h, y1_h)
    #plot!(x2_h, y2_h)
end

plot!()
```
